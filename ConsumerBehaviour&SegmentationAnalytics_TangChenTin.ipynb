## Preprocess Data

Before conducting cluster analysis, preprocessing the data is an essential step to ensure that the analysis is not skewed by the scale of the variables or by outliers. 

1. Dealing with Outliers: Outliers can significantly affect the clusters formed since clustering techniques often minimize the variance within a cluster. 
2. Standardization or Normalization: Since clustering algorithms such as K-means rely on distances between data points, it's important to scale features so that one feature doesn't dominate the distance calculation due to its scale. Standardization (Z-score normalization) transforms data such that each feature has a mean of 0 and standard deviation of 1. Normalization (min-max scaling) scales the data within a range, typically 0 to 1.


```python
import numpy as np
from scipy import stats
from sklearn.preprocessing import StandardScaler

# Assume 'data' is your DataFrame and you're interested in scaling 'age', 'income', 'purchase_frequency', and 'spending'

# Handling Outliers with IQR
Q1 = data[['age', 'income', 'purchase_frequency', 'spending']].quantile(0.25)
Q3 = data[['age', 'income', 'purchase_frequency', 'spending']].quantile(0.75)
IQR = Q3 - Q1

# Define a mask for outliers
outliers = ((data < (Q1 - 1.5 * IQR)) | (data > (Q3 + 1.5 * IQR))).any(axis=1)
data_no_outliers = data[~outliers]
data_no_outliers

# Seems like there isn't any outliers
```

```python
# Standardization
scaler = StandardScaler()
scaled_features = scaler.fit_transform(data_no_outliers[['age', 'income', 'purchase_frequency', 'spending']])

# Create a new DataFrame for the scaled features
cleaned_data = pd.DataFrame(scaled_features, columns=['age', 'income', 'purchase_frequency', 'spending'])

# Now 'cleaned_data' is ready for clustering
cleaned_data.describe()
```

## Cluster Analysis: 
To segment customers into distinct groups based on their characteristics. Methods like K-means or hierarchical clustering can be used here.

For cluster analysis, the K-means algorithm is a common starting point due to its simplicity and effectiveness:

1. Determine the Number of Clusters (k): Use the Elbow method to find a good value of k. You plot the sum of squared distances of samples to their closest cluster center, and where the plot starts to bend ("the elbow") suggests a good value for k.
2. Apply K-means Clustering: Once you've chosen k, you can apply the K-means algorithm to segment your data into k clusters.

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Assume 'cleaned_data' is your preprocessed DataFrame
features = cleaned_data[['age', 'income', 'purchase_frequency', 'spending']]

# Elbow Method to determine k
wcss = [] # Within-cluster sum of square
for i in range(1, 11): # Test for 1-10 clusters
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(features)
    wcss.append(kmeans.inertia_)

# Plot the Elbow graph
plt.figure(figsize=(10, 8))
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS') # Within-cluster sum of square
plt.show()
```

The Elbow Method graph is used to find the optimal number of clusters for K-means clustering. It shows the within-cluster sum of squares (WCSS) on the y-axis as a function of the number of clusters on the x-axis. WCSS is a measure of variance within each cluster, and the goal is to minimize this value, which indicates that the clusters are tight and well separated from each other.

As the number of clusters increases, the WCSS will naturally decrease since the points will be closer to the centroids they are assigned to. However, after a certain point, the decrease in WCSS will slow down significantly, indicating that adding more clusters doesn't provide much better modeling of the data. This point is known as the "elbow," and the number of clusters at this point is considered to be the optimal number.

In the graph, the WCSS decreases rapidly up to around 3 or 4 clusters, and then the rate of decrease slows down, suggesting that additional clusters beyond this point are not capturing as much new variance in the data.

While there is not a sharp distinct elbow in this plot, the curve starts to plateau somewhat after 3 clusters. Therefore, it's reasonable to choose 3 or 4 clusters as the optimal number for K-means algorithm. The exact choice might depend on additional domain knowledge or further analysis such as silhouette scores, but in general, based on this graph, I would suggest starting with 3 clusters and then examining the results to see if they make sense intuitively.

### Silhouette Coefficient

There are several metrics that can be used to compare the performance of K-means clustering with different numbers of clusters (k). Besides the Elbow method, the Silhouette Coefficient is a commonly used metric. The Silhouette Coefficient is a measure of how similar an object is to its own cluster (cohesion) compared to other clusters (separation). The Silhouette Score ranges from -1 to 1, where a high value indicates that the object is well matched to its own cluster and poorly matched to neighboring clusters. If the score is around 0, it indicates overlapping clusters.

```python
from sklearn.metrics import silhouette_score

# Assuming 'features' is your preprocessed data ready for clustering
silhouette_scores = []

# Range of k you want to test
k_values = range(2,6)

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=0)
    cluster_labels = kmeans.fit_predict(features)
    silhouette_avg = silhouette_score(features, cluster_labels)
    silhouette_scores.append(silhouette_avg)
    print(f"For n_clusters = {k}, the silhouette score is {silhouette_avg:.4f}")

# Plotting the silhouette scores
plt.figure(figsize=(10, 8))
plt.bar(k_values, silhouette_scores)
plt.title('Silhouette Scores for Different Numbers of Clusters')
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette Score')
plt.xticks(k_values)
plt.show()
```

### Interpretation:

- For n_clusters = 2: The score is 0.3060, which is the highest among the values you have provided. This suggests that when the dataset is partitioned into two clusters, the separation between the clusters is better than for higher numbers of clusters.
- For n_clusters = 3: The score drops slightly to 0.2653. This decrease indicates that the distinction between clusters is less clear than with two clusters, although the score still suggests a reasonable structure.
- For n_clusters = 4: The score further decreases to 0.2588, indicating that the additional cluster may be capturing some of the internal variances within a natural group, thereby making clusters less distinct.

##### Should consider their interpretability in a business context, how many k can result in more business marketing benefits

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import pandas as pd

# Assuming 'features' are the columns you want to use for clustering
features = cleaned_data[['age', 'income', 'purchase_frequency', 'spending']]

# Initialize the scaler
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

# Dictionary to hold the KMeans models and centroids
kmeans_models = {}
centroids = {}

for k in range(2, 5):  # Looping from 2 to 4 clusters
    kmeans = KMeans(n_clusters=k, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(scaled_features)
    cluster_labels = kmeans.predict(scaled_features)
    
    # Store the KMeans model
    kmeans_models[k] = kmeans
    
    # Compute and store the cluster centers
    centroids[k] = scaler.inverse_transform(kmeans.cluster_centers_)
    
    # Add cluster labels to the original DataFrame
    cleaned_data[f'cluster_k{k}'] = cluster_labels

# Now you have a dictionary of KMeans models and centroids for each k
# To access the model for k=3, you would use kmeans_models[3]
# To access the centroids for k=3, you would use centroids[3]

# Display the cluster centroids for k=3
print("Cluster Centers for k=3:")
print(pd.DataFrame(centroids[3], columns=features.columns))

# Display the first few rows of the DataFrame with the clusters
print("\nFirst few rows with cluster labels for k=2, k=3, and k=4:")
print(cleaned_data.head())
```
